```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
X,y = make_blobs(
    n_samples=1000,
    centers=3,
    n_features=2,
    random_state=42,
    
)
```

```python
X,y
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler
scalar =StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train_scaled = scalar.fit_transform(X_train)
X_test_scaled = scalar.transform(X_test)
```

* We only use X in Unsupervised ML

```python
from sklearn.cluster import KMeans

# Elbow method  to select K value
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(X_train_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
# Plot elbow curve
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of Clustest")
plt.ylabel('WSCC')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3,init='k-means++')
```

```python
kmeans.fit_predict(X_train_scaled)
```

```python
y_pred = kmeans.predict(X_test_scaled)
```

```python
y_pred
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_pred)
```

### **Validating the K value**
1. Knee Locator
2. Silhouette Scoring

```python
## 1. Kneelocator
!pip install kneed
```

```python
from kneed import KneeLocator

kl = KneeLocator(range(1,11),wcss,curve='convex',direction='decreasing')
```

```python
kl.elbow
```

```python
# 2. Silhouette Scoring
from sklearn.metrics import silhouette_score

silhoit_coefficients = []
for k in range(2,11):
    kmeans = KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(X_train_scaled)
    score=silhouette_score(X_train_scaled,kmeans.labels_)
    silhoit_coefficients.append(score)
```

```python
silhoit_coefficients
```

```python
# Plot elbow curve
plt.plot(range(2,11),silhoit_coefficients)
plt.xticks(range(2,11))
plt.xlabel("Number of Clustest")
plt.ylabel('WSCC')
plt.show()
```

For k = 3 the silhouette score is high
