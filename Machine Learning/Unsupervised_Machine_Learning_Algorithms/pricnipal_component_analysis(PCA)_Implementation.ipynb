```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd 
%matplotlib inline 
```

```python
# Load the dataset
from sklearn.datasets import load_breast_cancer

cancer_dataset = load_breast_cancer()
```

```python
cancer_dataset.keys()
```

```python
# All information aboy=ut the dataset
cancer_dataset.DESCR
```

```python
# Create dataframe
df = pd.DataFrame(cancer_dataset['data'],columns=cancer_dataset['feature_names'])
df.head()
```

* Aim is not to remove the features but to extract the feature which capture the maximum variance 


```python
# Standardization (mean=0,sd=1)
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler() 
scalar.fit(df)
scaled_data = scalar.transform(df)
```

```python
scaled_data
```

```python
# Applying PCA algorithm
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
data_pca = pca.fit_transform(scaled_data)

data_pca
```

```python
pca.explained_variance_
```

```python
pca=PCA()
pca.fit_transform(df)
pca.explained_variance_
```

```python
plt.figure(figsize=(10, 7))
# Scatter plot
scatter = plt.scatter(data_pca[:, 0], data_pca[:, 1], c=cancer_dataset['target'], cmap='plasma')
# Adding color bar to indicate target values
plt.colorbar(scatter, label='Target')
# Adding labels and title
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Cancer Dataset')

plt.show()
```
